```python
import pandas as pd

# Load the datasets
file1 = 'file:///Users/user/Desktop/airlinequality%20(3)%20copy%201.csv'
file2 = 'file:///Users/user/Desktop/airlinequality%202.csv'
file3 = 'file:///Users/user/Desktop/airlinequality%203%20.csv'

# Read the datasets
df1 = pd.read_csv(file1)
df2 = pd.read_csv(file2)
df3 = pd.read_csv(file3)

# Merge the datasets
df = pd.concat([df1, df2, df3], ignore_index=True)

# Save the merged dataset
df.to_csv('file:///Users/user/Desktop/airlinequality%20(3)%20copy%201.csv', index=False)
df.head()
```

```python
# Combine the datasets
combined_data = pd.concat([df1, df2, df3], ignore_index=True)

# Standardize column names: remove spaces and convert to lowercase
combined_data.columns = combined_data.columns.str.strip().str.lower().str.replace(' ', '_')

# Display combined data columns to identify and clean
combined_data.columns.tolist()

```

```python
# Identify key columns to retain
key_columns = [
    'text_sub_header',
    'text_sub_header_2',
    'text_sub_header_3',
    'text_content',
    'review-value',
    'review-value_2',
    'review-rating-header_12',
    'rating-10_2',
    'text_header'
]

# Drop irrelevant columns
df1 = combined_data[key_columns]

# Rename columns for clarity
df1.columns = [
    'Country',
    'Reviewer',
    'Date',
    'Content',
    'Traveller type',
    'Seat type',
    'Recommendation',
    'Rating',
    'Header'
]

# Display cleaned data info
df1.info()

# Display the first few rows of the cleaned data
df1.head()
```

```python
import numpy as np
df1['Recommendation'].replace(np.nan,'Not Recommended', inplace=True)
df1

```

```python

import re

# Clean specific columns
df1.loc[:, 'Header'] = df1['Header'].str.replace('"', '').str.replace('✅', '').str.replace('|', '')
df1.loc[:, 'Content'] = df1['Content'].str.replace('"', '').str.replace('✅', '').str.replace('|', '')
df1.head()
```

```python
import matplotlib.pyplot as plt
from wordcloud import WordCloud

# Summary statistics: average rating
average_rating = df1['Rating'].mean()

# Distribution of ratings
plt.figure(figsize=(10, 6))
df1['Rating'].value_counts().sort_index().plot(kind='bar')
plt.title('Distribution of Ratings')
plt.xlabel('Rating')
plt.ylabel('Frequency')
plt.grid(True)
plt.show()

# Common themes in reviews: generate word cloud for review content
text = ' '.join(df1['Content'])
wordcloud = WordCloud(width=800, height=400, background_color='white').generate(text)

# Display the word cloud
plt.figure(figsize=(10, 6))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis('off')
plt.title('Common Themes in Reviews')
plt.show()

average_rating
```
